### Some Questions
* Do students in more popular majors make more money?
* Do students that majored in subjects that were majority female make more money?
* Is there any link between the number of full-time employees and median salary?

```python
# let 'Total' be a substitute for the popularity of the major 
# in order to get an idea of more popular majors earn more
ax7 = recent_grads.plot(x='Total', y='Median',
                        kind='scatter',
                        title='Total vs Median');
```

```python
# Does the percent of women in a majority correlate with earnings?
ax8 = recent_grads.plot(x='ShareWomen', y='Median',
                        kind='scatter',
                        title='Share of Women vs Median');
```

```python
# Any link between number of full employment and earnings?
ax9 = recent_grads.plot(x='Full_time', y='Median',
                        kind='scatter',
                        title='Full Time vs Median');
```

### Perhaps a few Anwsers...
The only question we can answer at this time is whether majors with more women earn more, the answer is clearly no. From the graph `ax8` we can see a downward trend as the percent of women in the major increases. 

As for the other questions, there may be some relationship but there also appears to be a lot of noise and a lot of clustering around the origin. This makes it difficult to draw conclusions from the data.
